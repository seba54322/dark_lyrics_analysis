```python
# -*- coding: utf-8 -*-
#!/usr/bin/python
%matplotlib inline
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords 
import matplotlib.pyplot as plt
import time
```

# Preparar dataset total

## Abrir df_en

```python
df_en = pd.read_csv('./from_scratch/002_df_ingles_puro.csv')
df_en.head(100)
```

## Stopwords

```python
stopwords = ["now", "death", "life", "will"] + list(STOPWORDS)
stopwords = set(stopwords)
```

# Nubes dataset total


## Todas las letras, inglés, custom stopwords, top 200 df_en.lyric

```python
#Ingresa df subconjunto 
subconjunto = df_en.lyric
# Ingresa descripción de la visualización
registro = 'Todas las letras, inglés, custom stopwords, top 200'

##############################################

#Ruta
ruta = './from_scratch/nubes_de_palabras/'
start = time.time()
palabras = '' 
for val in subconjunto: 
    val = str(val) 
    tokens = val.split() 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower()
    palabras += " ".join(tokens)+" "

wordcloud = WordCloud(width = 1600, 
                      height = 1200, 
                      background_color ='black',
                      max_words=200,
                      stopwords = stopwords,
                      collocations=True,
                      relative_scaling=0.8,
                      normalize_plurals=True,
                      min_font_size = 10).generate(palabras)

end = time.time()
hours, rem = divmod(end-start, 3600)
minutes, seconds = divmod(rem, 60)
print('Time elapsed generating wordcloud:')
print("{:0>2}:{:0>2}:{:05.2f}".format(int(hours),int(minutes),seconds))

print (registro)                    
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()

wordcloud.to_file((ruta + registro + '.png'))
```
